### pie

```python
x=np.array([1,2,3,4])
plt.pie(x)
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple',"banana",'cherry','pineapple']
myexplod=[0,0,2,0]
plt.pie(x,labels=labels,explode=myexplod)    #for cut the part
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple',"banana",'cherry','pineapple']
myexplod=[0,0,1,0]
plt.pie(x,labels=labels,explode=myexplod,startangle=90,shadow=True)    
plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple',"banana",'cherry','pineapple']
myexplod=[0,0,1,0]
c=['red','yellow','pink','orange']
plt.pie(x,labels=labels,colors=c,explode=myexplod,startangle=90,shadow=True,autopct='%1.1f%%')    #for cut the part for 3d

plt.show()
```

```python
x=np.array([1,2,3,4])
labels=['apple',"banana",'cherry','pineapple']
myexplod=[0,0,1,0]
plt.pie(x,labels=labels,colors=c,explode=myexplod,startangle=90,shadow=True,autopct='%1.1f%%')
plt.legend(title="mypychart")
plt.show() 
```
